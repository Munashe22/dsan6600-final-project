```python
import pandas as pd
```

```python
enter_df = pd.read_csv('models_advanced/entrance_feature_importance_advanced.csv')
exit_df = pd.read_csv('models_advanced/exit_feature_importance_advanced.csv')
```

```python
most_important_enter = enter_df.sort_values(by="Normalized Importance").tail(10)

most_important_enter
```

```python
features_enter = ['Cam 2 Entropy','Cam 1 Car Count','Cam 1 Entropy','Cam 4 Entropy', 'Cam 3 Entropy','Cam 4 Embedding Std. Dev','Cam 4 Scene Complexity','Cam 4 Vehicles Exiting','Cam 3 Flow Rate','Cam 1 Vehicles in Middle Lane']

most_important_enter['features'] = features_enter
most_important_enter['importance'] = most_important_enter['Normalized Importance']*100

most_important_enter
```

```python
import matplotlib.pyplot as plt

plt.barh(most_important_enter["features"], most_important_enter["importance"])
plt.xlabel("Random Forest Feature Importance (%)")
plt.title("Feature Importances (Roundabout Enter)")
plt.tight_layout()
plt.savefig("plots/feature_importance_enter.png")
plt.show()
```

```python
most_important_exit = exit_df.sort_values(by="Normalized Importance").tail(10)
most_important_exit['importance'] = most_important_exit['Normalized Importance']*100


most_important_exit
```

```python
features_exit = ['Cam 3 Bus Count', 'Cam 3 Embedding Std. Dev.','Cam 1 Vehicles Exiting','Cam 3 Occupancy (Center Zone)',
                 'Cam 2 Entropy','Cam 4 Directional Flow (Upward)','Cam 1 Car Count','Cam 4 Vehicles Exiting','Cam 4 Directional Flow Left','Cam 4 Avg. Vehicle Speed']

most_important_exit['features'] = features_exit

most_important_exit
```

```python
import matplotlib.pyplot as plt

plt.barh(most_important_exit["features"], most_important_exit["importance"])
plt.xlabel("Random Forest Feature Importance (%)")
plt.title("Feature Importances (Roundabout Exit)")
plt.tight_layout()
plt.savefig("plots/feature_importance_exit.png")
plt.show()
```
